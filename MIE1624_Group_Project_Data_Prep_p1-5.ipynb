```python
import re
import nltk
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
nltk.download('punkt')
nltk.download('stopwords')
```

```python
skills_df=pd.read_csv('skills_book final.csv')
skills_df=skills_df.groupby('Umbrella_Term').Skills.apply(list).reset_index()
display(skills_df)
```

```python
dict_skills={}
for i in skills_df.index:
    dict_skills[skills_df.loc[i,'Umbrella_Term']]=skills_df.loc[i,'Skills']
display(dict_skills)
```

```python
def clean_text(text):
    #all_reviews = list()
    #lines = data["Description"].values.tolist()
    #for text in lines:
    text = text.lower()
    pattern = re.compile('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+')
    #text = pattern.sub('', text)
    text = re.sub(r"\n", " ", text)
    text = re.sub(r"[,.\"!@#$%^&*(){}?/;`~:<>+=-]", "", text)
    tokens = word_tokenize(text)
    table = str.maketrans('', '', string.punctuation)
    stripped = [w.translate(table) for w in tokens]
    words = [word for word in stripped if word.isalpha()]
    stop_words = set(stopwords.words("english"))
    stop_words.discard("not")
    #PS = PorterStemmer()
    #words = [PS.stem(w) for w in words if not w in stop_words]
    words = [w for w in words if not w in stop_words]
    words = ' '.join(words)
    #all_reviews.append(words)
    #return all_reviews
    return words  
```

```python
df['cleaned_description'] = df['Description'].apply(clean_text)
```

```python
df['cleaned_description'][0]
```

```python
def find_text(text, value):
    for i in value:
        pattern = re.compile(i)
        result = pattern.search(text)
        if result != None:
            return 1
        else:
            pass
    return 0
```

```python
for key, value in dict_skills.items():
    df[key] = df['cleaned_description'].apply(find_text, value=value)
display(df)
```
