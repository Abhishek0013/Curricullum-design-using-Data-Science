<h3> Select courses that has words data, science, analytics, etc </h3>

```python
#Extract relevant skills from tags for each course 
df.shape
s1 = ['Data','Analysis', 'Analytics', 'Machine', 'Regression', 'Python']
s2 = [s.lower() for s in s1]
search_values = s1 + s2
df1=df[ df['Tags'].str.contains('|'.join(search_values)) ]
df1.shape
```

```python
display(df1)
```

```python
l=[]
for i in df1.index:
    if df1.loc[i,'Skills']=='[]':
        l.append(i)
display(l)
```

```python
#Filtering only data science related courses
df1.drop(l,inplace=True)
display(df1)
```

<h3> Select the courses that has skills listed else (if NaN then) drop it</h3>

**To segregate all the skills taught is each course, we have to extarct all the skills from the tags columns for encoding. Courses without any information are not useful hence they are dropped.**

```python
#Drop courses with no skills mentioned

df1['Skills'] = df1['Skills'].str.strip("[]\'")
df1['Tags'] = df1['Tags'].str.strip("[]\'\'")

df1['Skills'] = df1['Skills'].str.replace("\'","")
df1['Tags'] = df1['Tags'].str.replace("\'","")

df1 = df1.drop(columns='type')
df1=df1[df1['Skills'].str.len()>1]
df1
```

<h3> Drop courses in which the 'Rating' is not listed or NaN </h3>

**Our approach is to search for top courses based on their Ratings.Hence courses without any Ratings have to be dropped as they are of no use.**

```python
#Drop Nan values under Ratings column
df1 = df1[df1['Rating'].notna()]
df1.shape
```

```python
#Creating a list of skills from the skills column
sk=[]
for i in df1['Skills'].values:
    sk.append(i)
sk_str=','.join(sk)
display(sk_str)
```

```python
skills=sk_str.split(',')
skills_list=list(set(skills))
display(skills_list)
```

```python
#Creating a fuzzy search algorithm for skills encoding
from fuzzywuzzy import fuzz
def label_skills(text, value):
    for i in value:
        ratio=fuzz.token_set_ratio(i,text)
    if ratio>=0.90:
        return 1
    else:
        return 0
```

```python
#Skills encoding for each course
for i in skills_list:
    df1[i] = df1['Skills'].apply(label_skills, value=i)
display(df1)
```

```python
df1[df1.columns[7:]].describe()
```

```python
#Creating a list of encoded skills with all 0 values for courses 
drop_list=[]
for i in df1.columns[7:]:
    if df1[i].mean()==0:
        drop_list.append(i)
display(drop_list)
```

```python
#Dropping encoded skills with all 0 values for courses
df2=df1.drop(drop_list,axis=1)
display(df2)
```

```python
df2[df2.columns[7:]].describe()
```

# Selection of top courses

**Here we trying to connect our results from the Indeed job posting analysis to use them to identify the top courses for most important skills for each job title.This in turn will be used to suggest chnages in the MIE1624 course curriculum and also in design process for a Masters in Data Science and AI program.**

```python
#Creating a function to display top courses for a particuler subject/skill
def top_courses(data,skill,n):
    course_id=[]
    for i in data.index:
        if data.loc[i,skill]==1:
            course_id.append(i)
    courses_df=data.loc[course_id,['title','url','Rating']]
    top_courses_df=courses_df.sort_values(by='Rating', ascending=False)[0:n]
    return top_courses_df
```

```python
top_courses(df2,skill='Machine Learning Concepts',n=7)
```
